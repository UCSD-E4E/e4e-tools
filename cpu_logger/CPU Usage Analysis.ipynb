```python
import datetime as dt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cpuLogFile = open('cpu_log.log', 'r')
```

```python
data = []
for line in cpuLogFile:
    timestamp = dt.datetime.fromtimestamp(int(line.split()[0]))
    userTime_ticks = int(line.split()[2])
    niceTime_ticks = int(line.split()[3])
    systemTime_ticks = int(line.split()[4])
    idleTime_ticks = int(line.split()[5])
    iowait_ticks = int(line.split()[6])
    irq_ticks = int(line.split()[7])
    softIRQ_ticks = int(line.split()[8])
    total_ticks = userTime_ticks + niceTime_ticks + systemTime_ticks + idleTime_ticks + iowait_ticks + irq_ticks + softIRQ_ticks
    data.append({
        'timestamp': timestamp,
        'userTime_ticks': userTime_ticks / total_ticks,
        'niceTime_ticks': niceTime_ticks / total_ticks,
        'systemTime_ticks': systemTime_ticks / total_ticks,
        'idleTime_ticks': idleTime_ticks / total_ticks,
        'iowait_ticks': iowait_ticks / total_ticks,
        'irq_ticks': irq_ticks / total_ticks,
        'softIRQ_ticks': softIRQ_ticks / total_ticks
    })
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.plot(df['timestamp'], df['userTime_ticks'], label='userTime_ticks')
plt.plot(df['timestamp'], df['niceTime_ticks'], label='niceTime_ticks')
plt.plot(df['timestamp'], df['systemTime_ticks'], label='systemTime_ticks')
plt.plot(df['timestamp'], df['idleTime_ticks'], label='idleTime_ticks')
plt.plot(df['timestamp'], df['iowait_ticks'], label='iowait_ticks')
plt.plot(df['timestamp'], df['irq_ticks'], label='irq_ticks')
plt.plot(df['timestamp'], df['softIRQ_ticks'], label='softIRQ_ticks')
plt.legend()
```
